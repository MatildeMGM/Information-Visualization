# Multiple Views

# Multiple superposed charts
We can make more than one chart and plot one over the other with the operator '+'.
We will overlap two charts.
To save some code, we first create a base chart, and reuse it to obtain modifications in a variable.
Finally, the two variables are plot together

```python
#@title Superposition
import altair as alt
from vega_datasets import data

source = data.jobs.url

base = alt.Chart(source).mark_line().encode(
    alt.X('year:O'),
    alt.Y('perc:Q', axis=alt.Axis(format='%')),
    color='sex:N'
)

ch1 = base.transform_filter(alt.datum.job == 'Welder')

ch2 = base.transform_filter(alt.datum.job == 'Weaver')

ch1 + ch2
```

These two charts share the same colors for men and women. We could do it differently by using two different palettes and making the palettes independent for the charts.

```python
#@title Superposition. Changing palettes
import altair as alt
from vega_datasets import data

source = data.jobs.url

base = alt.Chart(source).mark_line().encode(
    alt.X('year:O'),
    alt.Y('perc:Q', axis=alt.Axis(format='%')),
)

ch1 = base.transform_filter(alt.datum.job == 'Welder').encode(
    alt.Color('sex:N', scale = alt.Scale(scheme = 'dark2')))

ch2 = base.transform_filter(alt.datum.job == 'Weaver').encode(
    alt.Color('sex:N', scale = alt.Scale(scheme = 'accent')))

(ch1 + ch2).resolve_scale(color = 'independent')
```

But this does not help us understanding how the colors associate to the different jobs. This can be solved ensuring that the legend plots the title properly.

```python
#@title Superposition. Configuring legends
import altair as alt
from vega_datasets import data

source = data.jobs.url

base = alt.Chart(source).mark_line().encode(
    alt.X('year:O'),
    alt.Y('perc:Q', axis=alt.Axis(format='%')),
)

ch1 = base.transform_filter(alt.datum.job == 'Welder').encode(
    alt.Color('sex:N', scale = alt.Scale(scheme = 'accent'),
    legend = alt.Legend(title = 'Welder')))

ch2 = base.transform_filter(alt.datum.job == 'Weaver').encode(
    alt.Color('sex:N', scale = alt.Scale(scheme = 'category10'),
              legend = alt.Legend(title = 'Weaver')))

(ch1 + ch2).resolve_scale(color = 'independent')
```

# Adding labels

Labels can be added using superposition with the text mark

```python
#@title Pie chart example
import pandas as pd
import altair as alt

source = pd.DataFrame({"fruit": ['orange', 'pear', 'apple', 'lemon', 'plum'], "value": [4, 6, 10, 3, 7]})

ch1 = alt.Chart(source).mark_arc().encode(
    theta=alt.Theta("value"),
    order = alt.Order('value', sort = 'descending'),
    color="fruit:N"
)

ch2 = alt.Chart(source).mark_text(radius = 100).encode(
    theta=alt.Theta("value", stack = True),   # We need to stack here
    order = alt.Order('value', sort = 'descending'),
    text = 'value:N',
)

ch1 + ch2
```

# Reusing charts
Charts can be reused changing some parameters

```python
#@title Pie chart example
import pandas as pd
import altair as alt

source = pd.DataFrame({"fruit": ['orange', 'pear', 'apple', 'lemon', 'plum'], "value": [4, 6, 10, 3, 7]})

baseChart = alt.Chart(source).encode(    # We do not define the mark
    theta=alt.Theta("value", stack = True),
    order = alt.Order('value', sort = 'descending')
)

ch1 = baseChart.mark_arc().encode(
        color="fruit:N"
)

ch2 = baseChart.mark_text(radius = 100).encode(
    text = 'fruit:N',
)

ch1 + ch2
```

```python
#@title Multiple superposition example: Connected dot plot

import pandas as pd

data = {'Country': ['Lithuania', 'Cyprus', 'Romania', 'Sweden', 'Finland', 'Lithuania', 'Cyprus', 'Romania', 'Sweden', 'Finland'],
        'Smoking rate': [0.36, 0.52, 0.31, 0.07, 0.14, 0.13, 0.21, 0.08, 0.06, 0.11],
        'Gender': ['M', 'M','M', 'M','M', 'F','F', 'F', 'F', 'F']}

df = pd.DataFrame(data)

ch1 = alt.Chart(df).mark_circle(opacity = 1, size = 50, color = 'teal').encode(
    alt.X('Smoking rate:Q', scale=alt.Scale(domain=(0,0.6))),
    alt.Y('Country:N', sort='-x'),
).transform_filter(alt.datum.Gender == 'M').properties(title = 'Smoking population')

ch2 = alt.Chart(df).mark_circle(opacity = 1, size = 50, color = 'lightsalmon').encode(
    alt.X('Smoking rate:Q', scale=alt.Scale(domain=(0,0.6))),
    alt.Y('Country:N', sort='-x'),
).transform_filter(alt.datum.Gender == 'F').properties(title = 'Smoking population')

ch3 = alt.Chart(df).mark_line(color = 'gray').encode(
    alt.X('Smoking rate:Q'),
    alt.Y('Country:N'),
    detail = 'Country'
).properties(title = 'Smoking population')

#ch3
ch3 + ch1 + ch2
```

## Axes configuration

We can also make some changes to the axes:

- Axes labels can be formatted  removed

```python
#@title Axes configuration example
import altair as alt
import pandas as pd

data = pd.DataFrame({'Country': ['Germany', 'France', 'Italy', 'Spain',
                                 'Poland', 'Romania', 'Netherlands'],
        'Population (Millions)': [83.155, 67.657, 59.236, 47.399,
                                  37.840, 19.202, 17.475]})


alt.Chart(data).mark_bar().encode(
    alt.X('Population (Millions):Q',
          axis= alt.Axis(labels=False,               # Removing the axis labels
                         ticks = False),             # Removing the ticks
          title = ''),                               # Removing the axis title
    y = alt.Y('Country:N', title = 'Countries'),
).properties(
    title = 'Population'
).configure_axis(
    grid = False                                     # Removing the grid
).configure_view(
    strokeWidth = 0, strokeOpacity = 0)              # Removing the border
```

# Priestley chart

```python
import pandas as pd
import random

# The data represents some income ranges (min & max) per month for year 2024
# Create a list to store the data
data = []

# Define the months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Generate data for each month
for month in months:
  for i in range(2):  # Two entries per month
    income = 1000 + random.randint(500, 1500)
    data.append({'Month': month, 'Income': income, 'Year': 2024, 'Day': 1 })

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Create the Priestley chart (thick horizontal lines)
ch1 = alt.Chart(df).mark_bar().encode(
    alt.X('min(Income):Q', title = 'Income').scale(domain = (1000, 2800)),  # Income on the x-axis
    alt.X2('max(Income):Q'),
    alt.Y('Month:N').sort(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']),  # Month on the y-axis, nominal type for categorical order
    alt.Color('Month:N').scale(scheme = 'category20'),  # Color the bars by month
).properties(title='Priestley chart')

# Create the minimum labels
ch2Min = alt.Chart(df).mark_text(xOffset = -20).encode(
    alt.X('min(Income):Q'),
    alt.Y('Month:N').sort(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']),  # Month on the y-axis, nominal type for categorical order,
    alt.Text('min(Income):Q'),
)

# Create the maximum labels
ch2Max = alt.Chart(df).mark_text(xOffset = 20).encode(
    alt.X('max(Income):Q'),
    alt.Y('Month:N').sort(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']),  # Month on the y-axis, nominal type for categorical order,
    alt.Text('max(Income):Q'),
)

ch1 +  ch2Min + ch2Max
```

# Dot maps
These are maps with dots to identify some value over the geography

They are quite easy to create using altair

```python
#@title Airports in the US per state

import altair as alt
from vega_datasets import data

map = alt.topo_feature(data.us_10m.url, feature = 'states')

usChart = alt.Chart(map).mark_geoshape(
    fill = 'lightgray',
    stroke = 'white',
).properties(
    width = 500,
    height = 300
).project('albersUsa')

airports = data.airports.url

airportsMap = alt.Chart(airports, title = 'Number of airports in US'
).transform_aggregate(
    latitude='mean(latitude)',
    longitude='mean(longitude)',
    count='count()',
    groupby=['state']
).mark_circle().encode(
    longitude='longitude:Q',
    latitude='latitude:Q',
    size=alt.Size('count:Q', title ='Number of Airports'),
    color=alt.value('steelblue'),
    tooltip=['state:N','count:Q']
)

usChart + airportsMap
```

### Parallel coordinate plots

In this case, we need to normalize, because we cannot put different Y axes with different scales.

```python
import altair as alt
from vega_datasets import data
from altair import datum

source = data.cars()

alt.Chart(source).transform_window(
    index='count()'
).transform_fold(
    ['Weight_in_lbs', 'Miles_per_Gallon', 'Cylinders', 'Displacement', 'Horsepower', 'Acceleration'],
).transform_filter(
    'isValid(datum.value) && isFinite(datum.value)'  # Some Miles per Gallon and Horsepower are null
).transform_joinaggregate(
    min='min(value)',
    max='max(value)',
    groupby=['key']
).transform_calculate(
    minmax_value='(datum.value - datum.min) / (datum.max - datum.min)'
).mark_line().encode(
    x=alt.X('key:N', title=None, axis=alt.Axis(labelAngle = -45)),
    y=alt.Y('minmax_value:Q', title = 'Value'),
    color=alt.Color('Origin:N'),
    detail='index:N',
    opacity=alt.value(0.25)
).properties(width=500)
```

# Scatterplot with marginal histograms

```python
import altair as alt
from vega_datasets import data
import pandas as pd


cars = data.cars()

base = alt.Chart(cars)
base_bar = base.mark_bar(opacity=0.3, binSpacing=0)

xscale = alt.Scale(domain=(0, 240))
yscale = alt.Scale(domain=(0, 500))

points = base.mark_circle().encode(
    alt.X('Horsepower').scale(xscale),
    alt.Y('Displacement:Q').scale(yscale),
    alt.Color('Origin'),
)

top_hist = (
    base_bar
    .encode(
        alt.X('Horsepower:Q')
            .bin(maxbins=20,
                 extent=xscale.domain).stack(None).title(''),
        alt.Y('count()').stack(None).title(''),
        alt.Color('Origin:N'),
    )
    .properties(height = 60)
)

right_hist = (
    base_bar
    .encode(
        alt.Y('Displacement:Q')
            .bin(maxbins=20,
                 extent=yscale.domain).stack(None).title(''),
        alt.X('count()').stack(None).title(''),
        alt.Color('Origin:N'),
    ).properties(width = 60)
)

top_hist & (points | right_hist)

```
