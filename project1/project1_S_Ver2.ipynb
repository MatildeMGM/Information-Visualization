## Cleaning data

```python
#file = r"C:\Users\steff\OneDrive\Skrivebord\DTU\5. semester\Information visualization\Project 1\data for projekt 1\cruz_list.csv"
file1 = "cruz_list.csv"
file2 = "flagged_words_trump_admin.csv"
file3 = "nsf_terminations_airtable.csv"

# Ensure the file is not open in another program
print(os.path.abspath(file1))
data1 = pd.read_csv(file1)

print(os.path.abspath(file2))
data2 = pd.read_csv(file2)

print(os.path.abspath(file3))
data3 = pd.read_csv(file3)

data1
```

```python
data1[['grant_number', 'in_cruz_list']] = data1['grant_number;in_cruz_list'].str.split(';', expand=True)
data1['grant_id'] = data1['grant_number'].astype(int)
data1['in_cruz_list'] = data1['in_cruz_list'].map({'TRUE': True, 'FALSE': False})
data1 = data1.drop(columns=['grant_number','grant_number;in_cruz_list'])

data1
```

### Exercise 1

```python
# Count terminations by state
state_counts = (
    data3[data3['status'] == 0]['org_state']
    .value_counts()
    .reset_index()
)
state_counts.columns = ['state', 'Terminations']

# Select top 10 states
top10_states = state_counts.head(10)

import altair as alt

alt.Chart(top10_states).mark_bar(color='steelblue').encode(
    y=alt.Y('state:N', sort='-x', title='state'),
    x=alt.X('Terminations:Q', title='Number of Cancellations')
).properties(
    title='Top 10 States With Most Cancellations'
)
```

### Exercise 2

```python
# Count cancelled grants by institution (org_name)
institution_counts = (
    data3[data3['status'] == 0]['org_city']
    .value_counts()
    .reset_index()
)
institution_counts.columns = ['Institution', 'Cancelled Grants']

# Show top 10 institutions with most cancellations
top10_institutions = institution_counts.head(10)
top10_institutions

# Visualize
alt.Chart(top10_institutions).mark_bar(color='indianred').encode(
    y=alt.Y('Institution:N', sort='-x', title='Institution'),
    x=alt.X('Cancelled Grants:Q', title='Number of Cancelled Grants')
).properties(
    title='Top 10 Institutions With Most Cancelled Grants'
)
```
